```python
import numpy as np
from scipy.optimize import fsolve
def moment_of_inertia_two_parallelepipeds_z(
    density,
    lx,
    ly,
    lz,
    r,
):
    """
    Moment of inertia around the global z-axis for
    two identical rectangular parallelepipeds.

    Centers of mass:
        (+r, 0, 0)
        (-r, 0, 0)

    Parameters
    ----------
    density : float
        Density of parallelepipeds.

    lx, ly, lz : float
        Side lengths.

    r : float
        Distance of each center from origin along x.

    Returns
    -------
    Iz : float
        Total moment of inertia around z-axis.
    """
    mass = density * lx * ly * lz
    # Intrinsic inertia of one prism around its own COM
    Iz_cm = (mass / 12.0) * (lx**2 + ly**2)

    # Parallel axis theorem
    Iz_one = Iz_cm + mass * r**2

    # Two identical prisms
    Iz_total = 2.0 * Iz_one

    return Iz_total
```

```python
T = 83.1
density = 11.34e3  # kg/m**3
# Observation/target body
r = 37e-2  # m
lx = 14e-2  # m
ly = 4e-2  # m
lz = 7e-2  # m
I = moment_of_inertia_two_parallelepipeds_z(density, lx, ly, lz, r)
k = 4*np.pi**2 *I /T**2
print(k)
```

```python
k = 1e-4
def aceleracion(phi):
    phi = phi[0]
    pos = np.array([r*np.cos(phi), r*np.sin(phi), 0])
    mt = density *lx*ly*lz
    ms = density *5e-2*10e-2*15e-2
    # mt = density *4/3* np.pi*(5.8e-2/2)**3
    # ms = density *4/3* np.pi*(15e-2/2)**3
    d =  5e-2
    pos_masa = np.array([r, d, 0])
    G = 6.67430e-11
    F = G* mt*ms/(np.linalg.norm(pos_masa-pos))**3 * (pos_masa-pos)
    tau_g = np.cross(pos, F)
    tau_z_g = tau_g[2]
    return -k*phi + tau_z_g
fsolve(aceleracion, 0) * 180/np.pi
```
